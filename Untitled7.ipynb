```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
#  – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
#  – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
#  – list – команда, которая выведет список всех документов в формате passport "2207 876234" "Василий Гупкин"
#  – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. 
#  – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. 
#  – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.
#  – add – команда, которая добавит новый документ в каталог и в перечень полок,
    
   

def get_name_by_number (number): # возвращаем имя по номеру документа (1)
    for doc in documents:
        if doc['number'] == number:
            return doc['name']
    return 'Not found'  

def get_shelf_by_number (number):  #  полка по номеру документа (2)
    for shelf, docs in directories.items():
        for doc in docs:
            if doc == number:
                return shelf
    return 'Not found'
    
def print_doc_list ():   # печатаем тип номер и имя (3)
    for doc in documents:
        print ('{} "{}" "{}"'.format(doc['type'], doc['number'], doc['name']))

def move_doc (number, shelf):  # замена номера полки (используем (2))
    old_shelf =  get_shelf_by_number (number)
    if old_shelf == 'Not found':
        print ('Number not found')
        return
    if shelf not in directories:
        print ('Shelf not found')
        return
    directories[old_shelf].remove(number)
    directories[shelf].append(number)
    print('Successfully moved number')
    
def add_new_shelf ():           # добавление полки
        for shelf_n in directories.keys():
            if new_shelf == shelf_n:
                print ('The shelf already exist')
                return
        directories [new_shelf] = []
        return directories
        
def del_docs(number):
    for i, doc in enumerate(documents):
        if doc['number'] == number:
            documents.pop(i)
            #break
    for doc in documents:
        print ('{} "{}" "{}"'.format(doc['type'], doc['number'], doc['name'])) 

def del_doc (number):  # удаление документа с полки (используем (2))
    old_shelf =  get_shelf_by_number (number)
    if old_shelf == 'Not found':
        print ('Number not found')
        return
    directories[old_shelf].remove(number)
    print('Successfully delete number')
      
def add_doc (number): # добавление документа
    documents.append({'type': 'passport', 'number': (number), 'name': 'Ivan Ivanov'})
    for doc in documents:
        print ('{} "{}" "{}"'.format(doc['type'], doc['number'], doc['name']))    
    directories['3'].append(number)
    print ('  SHELFS:  ',(directories))
    return   
    
command = input ('Input command: ')   # запрос команд 
if command == 'people':
    user_input = input ('doc number: ')
    print (get_name_by_number (user_input))
elif command == 'shelf':
    user_input = input ('doc number: ')
    print (get_shelf_by_number (user_input))
elif command == 'list':
    print_doc_list ()
elif command == 'move':
    number = input ('doc number: ')
    shelf = input ('new shelf: ')
    move_doc(number, shelf)
    pprint(directories) 
elif command == 'add shelf':
    new_shelf = input ('New shelf number: ')
    add_new_shelf ()
    print (directories)
elif command == 'delete':
    number = input ('Delete doc number: ')
    del_docs (number)
    del_doc (number)
    print (directories)
elif command == 'add':
    number = input ('New doc number:')
    add_doc (number)
else:
    print ('INCORRECT COMMAND')
```
